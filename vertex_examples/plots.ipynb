```julia
using JLD2,Plots
```

```julia
@load "isovertex_0.0-to-100.0.jld2" isodataF
```

```julia
spins = Int.(getindex.(isodataF,1));
vals = getindex.(isodataF,2);
```

```julia
ramp2=[-(-1+0im)^0k*exp(-im*angle(vals[2])*k)*vals[k+1] for k in spins];
```

```julia
scatter(spins, abs.(ramp2),m=:c, yaxis=:log10,ms=:3)
plot!(spins[2:end], 0.00025(spins[2:end]).^-6,lw=:2)
```
