```julia
@load "isovertex_0.0-to-30.0.jld2" isodataF 
```

```julia
spins = Int.(getindex.(isodataF,1));
```

```julia
vals = getindex.(isodataF,2);
```

```julia
ramp2=[-(-1+0im)^0k*exp(-im*angle(vals[2])*k)*vals[k+1] for k in spins];
```

```julia
plot(spins, real(spins.^6 .* ramp2),m=:c)
plot!(spins, imag(spins.^6 .* ramp2))
```

```julia
scatter(spins, abs.(spins.^6 .* ramp2),leg=false)
```

```julia
scatter(spins, real.(spins.^6 .* ramp2),leg=false)
```
